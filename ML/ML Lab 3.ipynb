# Importing Data:

```python
from sklearn.utils import shuffle
import pandas as pd
import numpy as np
np.random.seed(42)
df = pd.read_csv('creditcard.csv')
df = df.sample(frac=1)
df
```

```python
df=df.sample(frac=1,random_state=42)
test=df.tail(56962)
```

```python
test
```

```python
df=df.drop(test.index)
```

```python
train=df.sample(frac=0.8,random_state=42)
validation=df.drop(train.index)
```

```python
validation
```

```python
#Data Cleaning
```

```python
for column in train.columns:
    train[column] =train[column]/ train[column].abs().max() #training data
```

```python
for column in test.columns:
    test[column] = test[column]  / test[column].abs().max() #testing data
```

```python
train
```

```python
test
```

```python
for column in validation.columns: #validation
    validation[column] = validation[column]/validation[column].abs().max()
```

```python
validation
```

```python
#Classification
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
```

```python
model = LogisticRegression()
model.fit(train.drop('Class',axis=1),train['Class'])
predictions = model.predict(test.drop('Class',axis=1))
```

```python
predictions
```

```python
print(classification_report(test['Class'], predictions))
```

```python
classifier = MLPClassifier(hidden_layer_sizes=(10,2), max_iter=100,activation = 'relu',solver='adam',random_state=42)
X_train=train.drop('Class',axis=1)
Y_train=train['Class']
classifier.fit(X_train, Y_train)
X_val=validation.drop('Class',axis=1)
y_val=validation['Class']
y_pred = classifier.predict(X_val)
y_pred
```

```python
def accuracy(confusion_matrix):
    diagonal_sum = confusion_matrix.trace()
    sum_of_all_elements = confusion_matrix.sum()
    return diagonal_sum / sum_of_all_elements
```

```python
confusionmatric = confusion_matrix(y_pred, y_val)
print("Accuracy of MLPClassifier : ''", accuracy(confusionmatric))
```

```python
confusionmatric
```

```python
precision=(confusionmatric[0][0])/(confusionmatric[0][0]+confusionmatric[0][1])
recall=(confusionmatric[0][0])/(confusionmatric[0][0]+confusionmatric[1][0])
print(precision)
print(recall)
```
